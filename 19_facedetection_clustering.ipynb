```python
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
```

```python
riot = cv2.imread('images/riot.jpg')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(riot)
```

```python
gray = cv2.cvtColor(riot, cv2.COLOR_BGRA2GRAY)
```

```python
plt.imshow(gray)
```

```python
faces = face_cascade.detectMultiScale(gray)
```

```python
faces
```

```python
for (x, y, w, h) in faces:
    img = cv2.rectangle(riot, (x, y), (x + w, y + h), (255, 0, 0), 5)
```

```python
plt.imshow(img)
```

```python
def face_detector(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGRA2GRAY)
    faces = face_cascade.detectMultiScale(gray)
    for (x, y, w, h) in faces:
        img = cv2.rectangle(img, (x, y), (x + w, y + h), (255, 0, 0), 5)
    return plt.imshow(img)
```

```python
crowd = cv2.imread('images/crowd_1.jpeg')
```

```python
face_detector(crowd)
```

```python
toons = cv2.imread('images/cartoon_crowd.jpeg')
```

```python
face_detector(toons)
```

```python
crowd2 = cv2.imread('images/crowd_2.jpeg')
```

```python
face_detector(crowd2)
```

```python
def cat_detector(catim):
    catscade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalcatface.xml')
    gray = cv2.cvtColor(catim, cv2.COLOR_BGRA2GRAY)
    faces = catscade.detectMultiScale(gray)
    for (x, y, w, h) in faces:
        img = cv2.rectangle(catim, (x, y), (x + w, y + h), (255, 0, 0), 5)
    return plt.imshow(img)
```

```python
cats = cv2.imread('images/gatos_crowd.jpeg')
```

```python
cat_detector(cats)
```
