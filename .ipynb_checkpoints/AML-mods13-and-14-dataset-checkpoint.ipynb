## Determining Drivers of Customer Satisfaction

**Background**

In the Spring of 2017, a disturbing video surfaced of Aviation Security Officers forcibly removing an injured passenger from [United Express Flight 3411](https://en.wikipedia.org/wiki/United_Express_Flight_3411_incident). The passenger in the video, Dr. David Dao Duy Anh, was physically removed from the aircraft after refusing to give up his seat.

Immediately following the incident, United Airlines leadership reacted poorly, causing a PR crisis and compounding an already negative public perception.

The aftermath of the incident ignited a strong effort within United to improve customer relations. The airline implemented a policy that tied customer satisfaction metrics into the compensation of every employee and broadly increased investment into customer satisfaction projects. One such investment was the creation of a dedicated customer satisfaction analytics team

One specific metric tracked closely by United leadership is customer Net Promoter Score (NPS). In an effort to improve this metric, the customer satisfaction analytics team first sought to identify the key factors influencing this metric. Regression models are an ideal choice for such a task due to the interpretable nature of the algorithms; however, the non-linear and collinear nature of data poses challenges.
