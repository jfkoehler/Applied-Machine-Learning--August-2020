```python
import pandas as pd
```

```python
reviews = pd.read_csv("https://s3.amazonaws.com/amazon-reviews-pds/tsv/sample_us.tsv", sep = '\t')
```

```python
reviews.head()
```

```python
rev_df = pd.pivot(reviews, index = 'product_title', columns = 'customer_id', values = 'star_rating')
```

```python
rev_df.shape
```

```python
rev_df.head()
```

```python
rev_df.fillna(0, inplace = True)
```

```python
rev_df.head()
```

```python
from scipy.sparse import csr_matrix
```

```python
rev_sparse = csr_matrix(rev_df)
```

```python
from sklearn.metrics import pairwise_distances
```

```python
dists = pairwise_distances(rev_sparse, metric = 'cosine')
```

```python
dist_rev_df = pd.DataFrame(dists, index = rev_df.index, columns = rev_df.index)
```

```python
dist_rev_df['Baby Einstein Octoplush'].nsmallest(10)
```

```python
def find_reviews():
    name = np.random.choice(rev_df.index)
    return dist_rev_df[name].nsmallest(10)
```

### Trying Surprise

- Load in data to dataframe
- Create trainset
- Use any algorithm's `.fit` method

```python
import surprise
```

```python
reviews = pd.read_csv("https://s3.amazonaws.com/amazon-reviews-pds/tsv/sample_us.tsv", sep = '\t')
```

```python
reviews.head()
```

```python
#userID, reviewID, rating
rev_df = reviews[['customer_id', 'review_id', 'star_rating']]
```

```python
rev_df.columns = ['userID', 'itemID', 'rating']
```

```python
reviews['star_rating'].describe()
```

```python
from surprise import Reader
from surprise import Dataset
from surprise.model_selection import cross_validate
```

```python
reader = Reader(rating_scale=(1, 5))
```

```python
data = Dataset.load_from_df(rev_df, reader)
```

```python
train = data.build_full_trainset()
```

```python
from surprise import SVD
```

```python
algo = SVD()
```

```python
cross_validate(algo, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
```

```python
algo.fit(train)
```
